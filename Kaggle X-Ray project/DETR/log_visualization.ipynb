```python
import pandas as pd 
logs = pd.read_csv('log.csv')
logs
```

```python
new_col = []
for i in range(len(logs.columns)):
    new_col.append(logs.columns[i].strip())

logs.columns = new_col
logs
```

```python
logs = logs.drop(['train_lr', 'n_parameters'], axis=1)
```

```python
logs_col = logs.columns
row, col = logs.shape[0], logs.shape[1]
logs.shape
```

```python
for i in range(row):
    for j in range(col):
        logs.iloc[i,j] = logs.iloc[i,j].split(': ')[1] 
```

```python
logs
```

```python
logs.columns
```

```python
epochs = logs['epoch'].values.astype(int).tolist()
train_losses = logs['train_loss'].values.astype(float).tolist()
test_losses = logs['test_loss'].values.astype(float).tolist()
```

```python
from matplotlib import pyplot as plt
import numpy as np 

plt.figure(1)
plt.plot(epochs, train_losses)
plt.xlabel('epoch')
plt.ylabel('train_loss')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)


plt.figure(2)
plt.plot(epochs, test_losses)
plt.xlabel('epoch')
plt.ylabel('val_loss')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

plt.show()
```
